Operações Aritméticas

```python
#soma
print(4+4)
#subtração
print(10-2)
#multiplicação  
print(5*2)
#divisão
print(16/2)
#potenciação    
print(2**3)
# Módulo
print(10%3)
```

Função Type

```python
print(type(10))
print(type(10.5))
print(type("Olá Mundo"))
```

Operações com Número Float

```python
print(7.4+5.4) 
print(4+4.4) 
print(4+4)
print(4/2) 
print(4//2)
print(7/2.0)
print(7//2.0)
```

Conversão

```python
print(float(9))
print(int(9.8))
print(str(10))
```

Hexadecimal e Binário

```python
print(hex(394))
print(bin(394))
print(hex(286))
print(bin(286))
```

Função abs,round e pow

```python
print(abs(-8))
print(abs(8))
print(pow(4,2))
print(pow(5,3))
print(round(3.14151922, 2))
```
